```python
from IPython.display import clear_output
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.io import show, output_notebook, push_notebook
output_notebook()
```

```python
def make_plot(src):  
    
    p1 = figure(title="Loss during epochs", plot_width = 400, plot_height=400)
    p1.line(x='epochs', y='train_loss', source=src, color='blue', legend='train')
    p1.line(x='epochs', y='valid_loss', source=src, color='red', legend='valid')

    p1.xaxis.axis_label = 'epoch'
    p1.yaxis.axis_label = 'loss'

    p2 = figure(title="Accuracy during epochs", plot_width = 400, plot_height=400)
    p2.line(x='epochs', y='train_acc', source=src, color='blue', legend='train')
    p2.line(x='epochs', y='valid_acc', source=src, color='red', legend='valid')

    p2.xaxis.axis_label = 'epoch'
    p2.yaxis.axis_label = 'accuracy'
    
    return p1, p2

def update_plot(train_loss, train_acc, valid_loss, valid_acc, handle):
    
    new_data = ColumnDataSource(
        dict(
            train_loss=train_loss, 
            train_acc=train_acc, 
            valid_loss=valid_loss, 
            valid_acc=valid_acc,
            epochs = list(range(1, len(valid_acc)+1))
        )
    )
    
    source.data.update(new_data.data)
    push_notebook(handle=handle)
```

```python
train_loss, train_acc, valid_loss, valid_acc = [],[], [], []

epochs = list(range(1, len(train_loss)+1))
source =  ColumnDataSource(data=dict(epochs=epochs, 
                                     train_loss=train_loss, 
                                     valid_loss=valid_loss, 
                                     train_acc=train_acc, 
                                     valid_acc=valid_acc))

p1, p2 = make_plot(src=source)

plots = show(row(p1, p2), notebook_handle=True)

import numpy as np

for i in range(1,4):   
    train_loss.append(np.random.randint(0, 10 + 1))
    train_acc.append(np.random.randint(0, 10 + 1))
    valid_loss.append(np.random.randint(0, 10 + 1))
    valid_acc.append(np.random.randint(0, 10 + 1))
    
    update_plot(i, train_loss, train_acc, valid_loss, valid_acc, plots)
```
